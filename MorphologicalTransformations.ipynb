```
# Boundary
def boundary(img, kernel):
  return img - cv2.morphologyEx(img, cv2.MORPH_ERODE, kernel)

kernel = np.ones((10,10))
out_img = boundary(images['circles'], kernel)

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (15,5))
ax1.imshow(out_img)
ax2.imshow(images['circles'])
ax3.imshow(shiftNscale(images['circles'] - out_img))

ax1.set_title('black hat')
ax2.set_title('original')
ax3.set_title('diff')
plt.show()
```

```
#Hole Filling 
def fill(img, point_coord, kernel = np.ones((3,3)), max_iter = 1000, verbose = True):
  X = np.zeros(img.shape, dtype='uint8')
  X[point_coord[0],point_coord[1]] = img.max() 
  for k in range(max_iter):
    if verbose:
      if k%10 == 0:
        print('iteration number: {}'.format(k))
        imshow(X)
    Xn = np.bitwise_and(cv2.morphologyEx(X, cv2.MORPH_DILATE, kernel), np.bitwise_not(img))
    if (Xn == X).all():
      print('Filled! number of iterations needed: {}'.format(k))
      imshow(X)
      break
    else: 
      X = Xn
```

```
N = 200
im = boundary(images['circles'], kernel)
print('boundary of circles')
fill(im, (105,105), verbose = False)

print('hole in the pcb image')
fill(images['pcb'], (50, 80), verbose = False)
```

```
#convex hulls 
im = images['pcb']
contours, hierarchy = cv2.findContours(im, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# create hull array for convex hull points
hull = []
# calculate points for each contour
for i in range(len(contours)):
    # creating convex hull object for each contour
    hull.append(cv2.convexHull(contours[i], False))

# create an empty black image
drawing = np.zeros((im.shape[0], im.shape[1], 3), np.uint8)
# draw contours and hull points
cv2.drawContours(drawing, hull, -1, (255), 1, 8)

f, (ax1, ax2) = plt.subplots(1,2, figsize = (10,5))
ax1.imshow(drawing)
ax2.imshow(im)

ax1.set_title('convex hull')
ax2.set_title('original')
plt.show()
```

```
#thinning:
def thin(A, B):
  return A - cv2.morphologyEx(A, cv2.MORPH_HITMISS, B)

#generate the struct elements
Bs = []
B1 = np.ones((3,3))
B1[0,:] = 0
Bs.append(B1)
B2 = B1.copy()
B2[:,-1] = 0
Bs.append(B2)
for i in range(3):
  B1 = np.rot90(B1)
  B2 = np.rot90(B2)
  Bs.append(B1)
  Bs.append(B2)
  
def thinning(A, Bs = Bs):
  for B in Bs:
    A = thin(A,B)
  return A
```

```
# thinning
in_img = images['pcb']
out_img = thinning(in_img)

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (15,5))
ax1.imshow(out_img)
ax2.imshow(in_img)
ax3.imshow(shiftNscale(in_img - out_img))

ax1.set_title('thinning')
ax2.set_title('original')
ax3.set_title('diff')
plt.show()
```

```
#thinning:
def thick(A, B):
  return A|cv2.morphologyEx(A, cv2.MORPH_HITMISS, B)

#generate the struct elements
Bs = []
B1 = np.ones((3,3))
B1[0,:] = 0
Bs.append(B1)
B2 = B1.copy()
B2[:,-1] = 0
Bs.append(B2)
for i in range(3):
  B1 = np.rot90(B1)
  B2 = np.rot90(B2)
  Bs.append(B1)
  Bs.append(B2)
  
def thickening(A, Bs = Bs):
  for B in Bs:
    A = thick(A,B)
  return A
```

```
# thickening
in_img = images['pcb']
out_img = thickening(in_img)

f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (15,5))
ax1.imshow(out_img)
ax2.imshow(in_img)
ax3.imshow(shiftNscale(in_img - out_img))

ax1.set_title('thickening')
ax2.set_title('original')
ax3.set_title('diff')
plt.show()
```

```
def skelly(A, B = cv2.getStructuringElement(cv2.MORPH_CROSS, (3,3)), K = 5):
  skel = np.zeros(A.shape, np.uint8)
  erroded = A
  
  while True:
    Sk = (erroded - cv2.morphologyEx(erroded, cv2.MORPH_OPEN, B))
    skel = np.bitwise_or(skel, Sk)
    erroded = (cv2.erode(erroded,B))
    if np.all(erroded==0):
      break 
  return skel
```

```
input_im = images['pcb']
out_im = skelly(input_im)

imshow(out_im)
```

```
Bs = []

B = np.zeros((3,3))
B[1,1] = 1
B[:,0] = 1
Bs.append(B)
B2 = np.zeros((3,3))
B2[0,0] = 1
B2[1,1] = 1
Bs.append(B2)
for i in range(3):
  B = np.rot90(B)
  B2 = np.rot90(B2)
  Bs.append(B)
  Bs.append(B2)

def prune(A, N = 3, B = Bs):
  X1 = A
  for n in range(N):
    X1 = thinning(X1,Bs = B)

  X2 = np.zeros(X1.shape, dtype='uint8')
  for b in Bs:
    X2 = np.bitwise_or(X2, cv2.morphologyEx(X1, cv2.MORPH_HITMISS, b))

  H = np.ones((3,3), dtype = 'uint8')
  X3 = X2
  for n in range(N):
    X3 = np.bitwise_and(cv2.dilate(X3,H), A)
  
  X4 = np.bitwise_or(X1, X3)
  return X4
```

```
blurred = cv2.GaussianBlur(images['morf_test'], (5, 5), 0)
edged = cv2.Canny(blurred, 50, 200, 255)

imshow(edged)
```

```
#https://stackoverflow.com/questions/53401617/finding-first-non-zero-value-in-an-image

#Expand edge
edged = cv2.copyMakeBorder(edged, 15, 15, 15, 15, cv2.BORDER_CONSTANT )

#get 4 sides
positions = np.nonzero(edged)
top = positions[0].min()
bottom = positions[0].max()
left = positions[1].min()
right = positions[1].max()

#show
rect = cv2.rectangle(cv2.cvtColor(edged, cv2.COLOR_GRAY2BGR)
    , (left, top), (right, bottom), (0,255,0), 1)

imshow(rect)
```

```
#narrow down the edges
s_range = 1
bdr = 15
tl = [np.nonzero(edged[top-s_range:top+s_range])[1].min()-bdr, top-bdr]
tr = [right+bdr, np.nonzero(edged[:,right-s_range:right+s_range])[0].max()-bdr]
bl = [left-bdr, np.nonzero(edged[:,left-s_range:left+s_range])[0].min()+bdr]
br = [np.nonzero(edged[bottom-s_range:bottom+s_range])[1].max()+bdr, bottom+bdr]
```

```
#some manual corrections bc the top didn't work perfectly
bl = [bl[0], bl[1]+15]
```

```
pts = np.array([tl,tr,br,bl], np.int32)
pts = pts.reshape((-1,1,2))
poly = edged.copy()
poly = cv2.polylines(poly,[pts],True, (255), 2)

imshow(poly)
```

```
#Warping to get the numbers more straight
height, width = edged.shape
pts1 = np.float32([tl, tr, br, bl])
pts2 = np.float32([[0,0], [width,0],[width, height], [0, height]])

mat = cv2.getPerspectiveTransform(pts1, pts2)
img = cv2.warpPerspective(edged, mat,(width, height))
imshow(img)
```

```
img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY)[1]
imshow(img)
```
